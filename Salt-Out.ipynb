```python
## USED POTENTIALS


sigma_ij = []
epsilon_ij = []
with open ('Potentials.txt') as f:
    for line in f:
        p = line.split()
        if len(p) !=0 and len(p) !=1:
            sigma_ij.append(float(p[0]))
            epsilon_ij.append(float(p[1]))


sigma_ij_system = []
epsilon_ij_system = []

for i in range(alpha*18-18,alpha*18):
    sigma_ij_system.append(sigma_ij[i])
    epsilon_ij_system.append(epsilon_ij[i])
    


sigma_ij_system_gas_water = []
epsilon_ij_system_gas_water = []

for i in range(0,6):
    sigma_ij_system_gas_water.append(sigma_ij_system[i])
    epsilon_ij_system_gas_water.append(epsilon_ij_system[i])


sigma_ij_system_gas_cation = []
epsilon_ij_system_gas_cation = []
for i in range(6,12):
    sigma_ij_system_gas_cation.append(sigma_ij_system[i])
    epsilon_ij_system_gas_cation.append(epsilon_ij_system[i])


sigma_ij_system_gas_anion = []
epsilon_ij_system_gas_anion = []
for i in range(12,18):
    sigma_ij_system_gas_anion.append(sigma_ij_system[i])
    epsilon_ij_system_gas_anion.append(epsilon_ij_system[i])
```

```python
## LJ calculation

LJ_interaction_gas_water = []

for i in range(len(sigma_ij_system_gas_water)):
    attraction = (np.pi*16*((epsilon_ij_system_gas_water[i]*sigma_ij_system_gas_water[i]**12)/9))/r_cut**9
    repulsion = -(np.pi*16*((epsilon_ij_system_gas_water[i]*sigma_ij_system_gas_water[i]**6)/3))/r_cut**3
    LJ_interaction_gas_water.append((attraction + repulsion)/T)


    
LJ_interaction_gas_cation = []

for i in range(len(sigma_ij_system_gas_cation)):
    attraction = (np.pi*16*((epsilon_ij_system_gas_cation[i]*sigma_ij_system_gas_cation[i]**12)/9))/r_cut**9
    repulsion = -(np.pi*16*((epsilon_ij_system_gas_cation[i]*sigma_ij_system_gas_cation[i]**6)/3))/r_cut**3
    LJ_interaction_gas_cation.append((attraction + repulsion)/T)


LJ_interaction_gas_anion = []

for i in range(len(sigma_ij_system_gas_anion)):
    attraction = (np.pi*16*((epsilon_ij_system_gas_anion[i]*sigma_ij_system_gas_anion[i]**12)/9))/r_cut**9
    repulsion = -(np.pi*16*((epsilon_ij_system_gas_anion[i]*sigma_ij_system_gas_anion[i]**6)/3))/r_cut**3
    LJ_interaction_gas_anion.append((attraction + repulsion)/T)

    
```

```python
# Density of ion pairs, cations,anions, water
rho_cation = []
rho_anion = []
rho_water = []
rho_ions = []
ion_concentration = []

for i in range(len(sym_particle_anion)):
    rho_cation.append((sym_particle_cation[i]*Mw_cation)/(Avo*0.0001))
    rho_anion.append((sym_particle_anion[i]*Mw_anion)/(Avo*0.0001))
    rho_ions.append(((sym_particle_cation[i]*Mw_cation)/(Avo*0.0001)) + ((sym_particle_anion[i]*Mw_anion)/(Avo*0.0001)))
    rho_water.append((sym_particle_water[i]*Mw_water)/(Avo*0.0001))
    ion_concentration.append((((sym_particle_cation[i]*Mw_cation)/(Avo*0.0001)) + ((sym_particle_anion[i]*Mw_anion)/(Avo*0.0001)))/Mw_ions)

```

```python
betta = np.empty([6,200])


for i in range(len(LJ_interaction_gas_water)):
    for j in range(len(sym_particle_water)):
        betta[i,j] = np.exp(-(sym_particle_water[j]*LJ_interaction_gas_water[i] + sym_particle_cation[j]*LJ_interaction_gas_cation[i] + sym_particle_anion[j]*LJ_interaction_gas_anion[i]))
```
